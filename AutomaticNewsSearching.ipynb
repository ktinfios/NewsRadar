```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
from newspaper import Article
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

import nltk 
nltk.download('popular')
```

```python
article_type = ["Investment / Funding", "Merger / Acquisition", "C-Level Management Change", "Product Launch"]
```

```python
def parse_article(article):
    article.download()
    article.parse()
    article.nlp()
    return {
        "title": article.title,
        "publish_date": article.publish_date,
        "url": article.source_url,
        "keywords": article.keywords,
        "summary": article.summary,
    }
```

```python
korber = Article(url="https://www.prnewswire.com/news-releases/korber-and-kkr-announce-the-strategic-acquisition-of-mercurygate-to-strengthen-their-supply-chain-software-business-302215435.html", language='en')
```

```python
infios = Article(url="https://www.infios.com/en/knowledge-center/news/edward-auriemma-becomes-new-ceo-of-business-unit-software", language='en')
```

```python
response = requests.get(articles[0])
print("Final URL after redirects:", response.url)
print("Status code:", response.status_code)
```

```python
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

# Open the URL
driver.get(articles[0])

# Get the final redirected URL
final_url = driver.current_url
print("Redirected to:", final_url)

driver.quit()
```
